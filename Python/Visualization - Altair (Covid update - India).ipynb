```python
import pandas as pd
import altair as alt
import datapane as dp
import numpy as np
```

```python
df = pd.read_csv('https://api.covid19india.org/csv/latest/states.csv', index_col = ['Date'], parse_dates=True)
df.head()
```

```python
# Unique states
np.unique(df.State)
```

```python
states = ['Delhi', 'Tamil Nadu', 'Maharashtra']

df_states = df[df.State.isin(states)]
```

```python
df.shape, df_states.shape
```

```python
g1 = alt.Chart(data = df_states.reset_index()).mark_line(size = 3).encode(
    x = 'Date:T',
    y = 'Confirmed',
    color = 'State',
).interactive()
```

```python
g2 = alt.Chart(df_states.reset_index()).mark_point(
    size=5, opacity = 0.7).encode(
    x = 'Date:T',
    y = 'Deceased',
    color = 'State'
    ).interactive()
g2
```

```python
g3 = alt.Chart(df_states).mark_boxplot(size = 50).encode(
    x = 'State:N',
    y = 'Deceased:Q',
    color = 'State'
).properties(width = 500)
g3
```

```python
plots = [g1, g2, g3]
```

```python
report = dp.Report(
    dp.Text(f"## Covid update in {', '.join(states)}"),
    dp.Group(*plots[:3], columns=2),
    #plots[2],
    dp.DataTable(df_states, caption=f'Dataset for {states}'),
)
```

```python
report.publish(name='Covid India Update', open=True,
               description="Covid report, using data from covid19india.org")
```
